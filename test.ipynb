```python
import re
from typing import Dict, Optional, Any
from rapidfuzz import process, fuzz

# ---------------- Data Mappings ---------------- #
UAE_IDENTIFIERS = {
    # Abu Dhabi
    **{str(n): "Abu Dhabi" for n in [1, 50] + list(range(4, 22))},

    # Dubai
    **{c: "Dubai" for c in list("ABCDEFGHJKLMNPRSTUVWXYZ")},
    **{cc: "Dubai" for cc in ["AA", "BB", "CC", "DD"]},

    # Sharjah
    **{str(n): "Sharjah" for n in range(1, 5)},

    # Ajman
    **{c: "Ajman" for c in list("ABCDEH")},

    # Fujairah
    **{c: "Fujairah" for c in list("ABCDEFGKMPRST")},

    # Ras Al Khaimah
    **{c: "Ras Al Khaimah" for c in ["A", "C", "D", "I", "K", "M", "N", "S", "V", "Y"]},

    # Umm Al Quwain
    **{c: "Umm Al Quwain" for c in list("ABCDEFGHIX")},
}

UAE_REGIONS_FULL = {
    "ABU DHABI": "Abu Dhabi",
    "DUBAI": "Dubai",
    "SHARJAH": "Sharjah",
    "AJMAN": "Ajman",
    "FUJAIRAH": "Fujairah",
    "RAS AL KHAIMAH": "Ras Al Khaimah",
    "UMM AL QUWAIN": "Umm Al Quwain",
    "AD": "Abu Dhabi",
    "RAK": "Ras Al Khaimah",
    "UAQ": "Umm Al Quwain"
}

UAE_REGIONS = list(UAE_REGIONS_FULL.values())


def autocorrect_region(token: str, threshold: int = 70) -> str:
    """Correct OCR token to closest UAE region name."""
    token = token.strip().title()
    match, score, _ = process.extractOne(token, UAE_REGIONS, scorer=fuzz.token_sort_ratio)
    return match if score >= threshold else token


def format_license_plate(data: Dict[str, Any]) -> Dict[str, Optional[str]]:
    """Extracts and formats UAE license plate data from OCR result."""
    rec_texts = [t.strip().upper() for t in data.get("rec_texts", []) if isinstance(t, str) and t.strip()]
    rec_texts = [autocorrect_region(t).upper() for t in rec_texts]

    region, category, number = None, None, None

    # 1. Match full emirate name
    for text in rec_texts:
        if text in UAE_REGIONS_FULL:
            region = UAE_REGIONS_FULL[text]
            break

    # 2. If not found, check identifier mapping
    if not region:
        for text in rec_texts:
            if text in UAE_IDENTIFIERS:
                region = UAE_IDENTIFIERS[text]
                category = text
                break

    # 3. Extract longest number (1–5 digits)
    for text in rec_texts:
        if re.fullmatch(r"\d{1,5}", text):
            if not number or len(text) > len(number):
                number = text

    # 4. Extract category if still missing
    if not category:
        for text in rec_texts:
            if text != number and re.fullmatch(r"(?:[A-Z]{1,2}|\d{1,2})", text):
                category = text
                break

    # 5. Build license plate string
    license_plate = None
    if category and region and number:
        license_plate = f"{category} {region} {number}"
    elif region and number:
        license_plate = f"{region} {number}"

    return {"license plate": license_plate, "category": category, "region": region, "number": number}


def process_json_data(ocr_data: dict) -> Dict[str, Optional[str]]:
    """Processes OCR data from JSON/dict and returns formatted result."""
    return format_license_plate(ocr_data)
```

```python
import json
import re
from typing import Dict, Optional, Any
from rapidfuzz import process, fuzz
from functools import lru_cache

# ---------------- Data Mappings ---------------- #
UAE_IDENTIFIERS = {
    # Example filled if needed...
}

UAE_REGIONS_FULL = {
    "ABU DHABI": "Abu Dhabi",
    "DUBAI": "Dubai",
    "SHARJAH": "Sharjah",
    "AJMAN": "Ajman",
    "FUJAIRAH": "Fujairah",
    "RAS AL KHAIMAH": "Ras Al Khaimah",
    "UMM AL QUWAIN": "Umm Al Quwain",
    "AD": "Abu Dhabi",
    "RAK": "Ras Al Khaimah",
    "SHJ": "Sharjah",
    "UAQ": "Umm Al Quwain"
}

# ---------------- Utilities ---------------- #
@lru_cache(maxsize=512)  # cache results for repeated tokens
def autocorrect_region(token: str, threshold: int = 70) -> str:
    """Autocorrect a token to closest UAE region using fuzzy matching (cached)."""
    token_clean = token.strip().title()
    if token_clean in UAE_REGIONS_FULL:  # exact match fast path
        return UAE_REGIONS_FULL[token_clean]

    match, score, _ = process.extractOne(
        token_clean, UAE_REGIONS_FULL.values(), scorer=fuzz.token_sort_ratio
    )
    return match if score >= threshold else token


def split_alphanum(token: str):
    """Splits token into components: handles abbreviations, dots, letters+digits."""
    token = autocorrect_region(token).upper()

    # Abbreviations with dots (U.A.E, Ρ.E.1) → keep whole
    if re.match(r"^([A-ZΑ-ΩΡ]\.){2,}[A-Z0-9]*$", token):
        return [token]

    # Normal '.' split
    if "." in token:
        return [p for p in token.split(".") if p]

    # Letters+digits (DD50748 → DD, 50748)
    m = re.match(r"^([A-Z]+)(\d+)$", token)
    if m:
        return [m.group(1), m.group(2)]

    return [token]


def extract_category(texts, region, number):
    """Extracts vehicle category code based on region-specific rules."""
    for text in texts:
        if text == number:
            continue
        if region == "Abu Dhabi" and (text == "AD" or (text.isdigit() and (text == "1" or 4 <= int(text) <= 21 or text == "50"))):
            return text
        if region == "Dubai" and re.fullmatch(r"[A-Z]{1,2}", text) and text not in {"I", "O", "Q"}:
            return text
        if region == "Sharjah" and text.isdigit() and 1 <= int(text) <= 5:
            return text
        if region == "Ajman" and text in {"A", "B", "C", "D", "E", "H"}:
            return text
        if region == "Ras Al Khaimah" and text in {"A", "C", "D", "I", "K", "M", "N", "S", "V", "Y"}:
            return text
        if region == "Fujairah" and text in {"A", "B", "C", "D", "E", "F", "G", "K", "M", "P", "R", "S", "T"}:
            return text
        if region == "Umm Al Quwain" and text in {"A", "B", "C", "D", "E", "F", "G", "H", "I", "X"}:
            return text
    return None


def format_license_plate(data: Dict[str, Any]) -> Dict[str, Optional[str]]:
    """Extracts and formats UAE license plate data from OCR result (optimized)."""
    rec_texts = [
        part
        for t in data.get("rec_texts", [])
        if isinstance(t, str) and t.strip()
        for word in t.strip().upper().split()
        for part in split_alphanum(word)
        if part
    ]

    print("Texts:", rec_texts)

    region_name, number, category = None, None, None
    max_len = 0

    # Single loop → extract region, number, and potential category
    for text in rec_texts:
        if not region_name and text in UAE_REGIONS_FULL:
            region_name = UAE_REGIONS_FULL[text]

        if re.fullmatch(r"\d{1,5}", text):  # number candidate
            if len(text) >= max_len:
                number, max_len = text, len(text)

    if region_name and not category:
        category = extract_category(rec_texts, region_name, number)

    license_plate = None
    if category and region_name and number:
        license_plate = f"{category} {region_name} {number}"
    elif region_name and number:
        license_plate = f"{region_name} {number}"

    return {"license plate": license_plate, "category": category, "region": region_name, "number": number}


def process_json_data(ocr_data: dict) -> Dict[str, Optional[str]]:
    return format_license_plate(ocr_data)
```
